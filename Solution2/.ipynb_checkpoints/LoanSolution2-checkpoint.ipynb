```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
train_df = pd.read_csv('data_cred_train.csv')
```

```python
train_df.head()
```

```python
train_df.shape
```

```python
train_df.describe().round(3)
```

```python
for i in train_df.columns:
    prc = np.mean(train_df[i].isnull())
    print(i, '- %.2f'%(prc * 100), 'perc', sep=' ')
```

```python
#Удалим строки, не содержащие никакую информацию
train_df = train_df.dropna(how='all')
```

```python
train_df['Credit Score'].describe().round(3)
```

```python
pd.qcut(train_df['Credit Score'],8).value_counts()
```

```python
#Избавимся от строк-выбросов в значениях кредитного рейтинга
train_df = train_df.drop(train_df[train_df['Credit Score'] > 1000].index)
pd.qcut(train_df['Credit Score'],8).value_counts()
```

```python
train_df.shape
```

```python
train_df['Credit Score'].isna().sum()
```

```python
train_df = train_df[train_df['Credit Score'].notna()]
```

```python
train_df['Credit Score'].describe().round(3)
```

```python
pd.qcut(train_df['Credit Score'],8).value_counts()
```

```python
train_df['Credit Allowed'] = 1
train_df.head()
```

```python
train_df['Chances Of Payment'] = train_df['Credit Score'] / 1000.0
train_df.head()
```
